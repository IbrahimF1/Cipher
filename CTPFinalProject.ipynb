```python
# Heatmap of Source and Destination IP Interactions
ip_interactions = df.pivot_table(index='SourceIP', columns='DestinationIP', values='DnsQuery', aggfunc='count').fillna(0)

# Plotting here (NEED TO FIX)
plt.figure(figsize=(12, 8))
sns.heatmap(ip_interactions, cmap="YlGnBu", annot=True, fmt=".0f", cbar_kws={'label': 'Number of Queries'})
plt.title("Heatmap of DNS Queries Between Source and Destination IPs")
plt.xlabel("Destination IP")
plt.ylabel("Source IP")
plt.show()
```

```python
# Pie Chart of Response Codes

response_code_counts = df['DnsResponseCode'].value_counts()

#WHAT THE CHART MEANS
# RCODE 0: No Error (Success)
# RCODE 3: Name Error (NXDOMAIN) (Non-Existent Domain)

# Set up labels in array
legend_labels = ["0: No Error (Success)", "3: Name Error (NXDOMAIN)"]
colors = ["#66c2a5", "#FF0000"]

# Plotting
plt.figure(figsize=(10, 10))
plt.pie(response_code_counts, labels=response_code_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Distribution of DNS Response Codes")
plt.legend(legend_labels, loc="best", title="Response Code Descriptions")
plt.show()
```
